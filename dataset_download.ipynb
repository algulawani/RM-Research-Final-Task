```python
import os

label_map = {
    '1': 'Surprise',
    '2': 'Fear',
    '3': 'Disgust',
    '4': 'Happiness',
    '5': 'Sadness',
    '6': 'Anger',
    '7': 'Neutral'
}

for split in ['train', 'test']:
    print(f"\n{'='*40}")
    print(f"  {split.upper()} SET")
    print(f"{'='*40}")
    total = 0
    split_path = os.path.join(path, 'DATASET', split)
    for folder in sorted(os.listdir(split_path)):
        folder_path = os.path.join(split_path, folder)
        count = len([f for f in os.listdir(folder_path) if f.endswith('.jpg')])
        total += count
        emotion = label_map.get(folder, 'Unknown')
        print(f"  {folder} ({emotion:>10}): {count:>5} images")
    print(f"  {'─'*30}")
    print(f"  {'TOTAL':>14}: {total:>5} images")
```
